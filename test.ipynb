```python
#!/usr/bin/env python
# coding: utf-8

# In[ ]:

import load_dataset
import user_stat

#User Interface function
#This function fetches the options and user_id and transaction id from the user

def test_data():
    try:
        f1='\033[1m'
        f2='\033[0m'
        color='\033[32m'+'\033[1m'
        number=0
        print("-----------------------------------------------------------")
        print(f1 +"WELCOME TO FRAUD DETECTION SYSTEM"+f2)
        print("-----------------------------------------------------------")
        #Loop for continous execution till exit
        while (number !=9):
            #Fetching inputs from user
            print(color+"\n Please select any of the below options \n 1.Minmax \n 2.Location Centroid \n 3.Distance between centroid and a transaction \n 4.Standard deviation \n 5.Variance \n 6.Fraudulent \n 7.Distance between 2 transaction of same user \n 8.Distance between two transaction of different users \n 9.Press 9 to logout\n" +f2)
            option=int(input(f1+"Please select one option  :  "+f2))
            if (option>0 and option <=8):                
                transaction_dataset ={}
                #Calling load_dataset to get the transaction data
                transaction_dataset=load_dataset.transdata_load()
                #Verify that the given user is valid
                uid=input(f1+"Enter your user_id please :  \n"+f2)
                verified_uid=user_stat.verify_user_id(uid,transaction_dataset)
                if (verified_uid != False):                    
                    user_selection(option,verified_uid,transaction_dataset) 
                else:
                    print(f1+"Invalid User !!"+f2)
            elif(option == 9):
                #To exit from the execution
                print(f1+"You are successfully logout from the system"+f2)
                break
            else:
                #Error message for wrong option selection
                print(f1+"Invalid options!Please provide the right option"+f2)  
    except ValueError:
        print(f1+"Invalid options! Please provide the right option"+f2)

#Calling function
#This function calls the function corresponding to the options selected and fetched next level of inputs from user
def user_selection(option,uid,transaction_dataset): 
    f1='\033[1m'
    f2='\033[0m'
    if option == 1:   
        #Calling Minmax function
        user_stat.minmax(uid,transaction_dataset)
    elif option == 2:
        #Calling centroid function
        user_stat.location_centroid(uid,transaction_dataset)
    elif option ==3:
        #Calling find_transaction fn to display all transaction details under the given user
        user_stat.find_transaction_desc(uid,transaction_dataset)
        desc1=input(f1+"Please select the one of the transaction details:(Input transaction id)  "+f2)  
        #Calling distance function
        user_stat.distance(uid,desc1,transaction_dataset)
    elif option ==4:
        #Calling standard deviation&Variance function
        sd=user_stat.standard_deviation_and_variance(uid,transaction_dataset)
        print(f1+"The standard_deviation"+f2,sd[0])
    elif option ==5:
        #Calling standard deviation&Variance function
        va=user_stat.standard_deviation_and_variance(uid,transaction_dataset)
        print(f1+"The variance is "+f2,va[1])
    elif option ==6:
        #Calling find_transaction fn to display all transaction details under the given user
        user_stat.find_transaction_desc(uid,transaction_dataset)
        desc=input(f1+"Please select the one of the transaction details:(Input transaction id)"+f2)
        #Calling fraud function
        user_stat.fradualent(uid,desc,transaction_dataset) 
    elif option ==7:
        #Calling find_transaction fn to display all trasaction details under the given user
        user_stat.find_transaction_desc(uid,transaction_dataset)
        desc1=input(f1+"Please select the one of the transaction details:(Input transaction id)"+f2)
        #Verify transaction id
        verify1=user_stat.verify_transaction_id(uid,desc1,transaction_dataset)
        desc2=input(f1+"Please select the one of the transaction details:(Input transaction id)"+f2)
        #Verify transaction id
        verify2=user_stat.verify_transaction_id(uid,desc2,transaction_dataset)
        if (verify1 == False or verify2 == False):
             print(f1+"Invalid transaction id"+f2) 
        else:
            #Calling the distance calculation function
            user_stat.distance_trans_userid(uid,desc1,desc2,transaction_dataset)
    elif option ==8:
        #Calling find_transaction function to display all transaction details under the given user
        user_stat.find_transaction_desc(uid,transaction_dataset)
        desc1=input(f1+"Please select the one of the transaction details:(Input transaction id)"+f2)
        #Verify transaction id
        verify=user_stat.verify_transaction_id(uid,desc1,transaction_dataset)        
        if (verify ==False):
            print(f1+"This trasaction id not exist"+f2)
        else:
            uid2=input(f1+"Enter second user_id please :  "+f2)
            #Verify that the user id is different
            if(uid ==uid2):
                print(f1+"Please give different user details..."+f2)                             
            user_stat.find_transaction_desc(uid2,transaction_dataset)
            desc2=input(f1+"Please select the one of the transaction details:(Input transaction id) : "+f2)
            #Verify transaction id
            verify=user_stat.verify_transaction_id(uid2,desc2,transaction_dataset)
            if (verify == False):
                print(f1+"This trasaction id not exist"+f2)
            else:
                #Calling distance calculation function
                user_stat.distance_userid(uid,uid2,desc1,desc2,transaction_dataset)
        
def main():
    test_data()
main()
    
```
